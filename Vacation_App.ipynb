```python
import seaborn as sns # For  plotting
import pandas as pd # For data processing
import numpy as np # For linear algebra
import pandas.testing as tm # For testing
import matplotlib.pyplot as plt # For plotting
from pandas.plotting import parallel_coordinates # For importing a module called "parallel_coordinates" from a library called "pandas" for creating parallel coordinates.
```

```python
# Read the dataset into a Pandas DataFrame
M6 = "15 - Vacations App - M6 - Relationships.xlsx"  #Assign the file path to a variable
df = pd.read_excel(M6) #Parse the M6 dataset into dataframe object df
df
```

```python
# Define a dictionary for questions and labels for the questions
questions = {
    'Q21': 'How likely are you to use a Vacations App based on this factor: Free',
    'Q22': 'How likely are you to use a new Vacations App based on this factor: Breath of Vacation locations',
    'Q23': 'How likely are you to use a new Vacations App based on this factor: Ensures privacy',
    'Q24': 'How likely are you to use a new Vacations App based on this factor: Protects identity',
    'Q25': 'How likely are you to use a new Vacations App based on this factor: Backups to cloud',
}
```

```python
# Parallel Coordinates Plot
plt.figure(figsize=(10, 6)) #create the figure for the parallel plot
custom_x_labels = ['App has breath of\n Vacation locations','App ensures Privacy', 'App protects Identity'] #Create custom labels for X axis
custom_y_labels = ['I will not use', 'Highly Unlikely', 'Unlikely', 'Unsure', 'Likely', 'Highly Likely'] #Create custom labels for Y axis
label_mapping = {} #Create a dictionary for custom labels for the legend.

# Assign custom labels to legend
for i, num_label in enumerate([-3, -2, -1, 0, 1, 2]):
    label_mapping[num_label] = custom_y_labels[i]

# Replace numerical labels with custom labels in the legend
legend_labels = [label_mapping[label] for label in [-3, -2, -1, 0, 1, 2]]

parallel_coordinates(df[['Q21', 'Q22', 'Q23', 'Q24']], 'Q21', colormap='viridis') #Generate the parallel coordinates plot
plt.title('Parallel Coordinates Plot')#Set the title
plt.legend(legend_labels,loc='center left', bbox_to_anchor=(1.0, 0.5)) #Show the legend
plt.xticks(range(3), custom_x_labels)#Display the custom x labels
plt.yticks([-3, -2, -1, 0, 1, 2], custom_y_labels)#Display the custom y labels
plt.savefig("PC.png")
plt.show() #Display the plot
```

```python
# Radar Graph
attributes = ['Q21', 'Q22', 'Q23', 'Q24', 'Q25'] #Set the attributes for the radar plot
attribute_labels=['App is free',
                  'App has breath \nof vacation locations',
                  'App ensures\n Privacy',
                  'App protects\n identity',
                  'App backs up to cloud']#Set custom attributes to display on the radar plot
data = df[attributes].mean().tolist() #Calculate mean values of the attributes
angles = [n / float(len(attributes)) * 2 * 3.14159265359 for n in range(len(attributes))] #Set the angles at which data points are plotted
angles += angles[:1]
data += data[:1]

plt.figure(figsize=(10, 10)) #create the figure for the parallel plot
plt.polar(angles, data, marker='o') #Create the polar plot
plt.fill(angles, data, 'b', alpha=0.1) #Fill the area under the radar graph
plt.xticks(angles[:-1], attribute_labels) #Set the custom x labels in the radar plot
plt.tick_params(axis='x', pad=20) # Set the padding to have space between the labels and the radar graph
plt.title('Radar Graph between Mean Values of different features of the Vacations App')#Set the title
plt.savefig("RC.png")
plt.show() #Display the plot
```
